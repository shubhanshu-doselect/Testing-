```python
import pandas as pd 
import mysql.connector 
import re 
import nltk 
from nltk.corpus import stopwords 
nltk.download('punkt') 
from nltk.tokenize import word_tokenize 
from collections import defaultdict 
from pysimilar import compare 
pd.options.display.max_colwidth = 100000
```

```python
def makeConnectSQL():
    mydb = mysql.connector.connect(user='root', password='root',host='127.0.0.1', database='Doselect',auth_plugin='mysql_native_password')
    return mydb

def getdata():
    mydb = makeConnectSQL()
    mycursor = mydb.cursor()
    mycursor.execute("select * from MCQData;")
    ls = mycursor.fetchall()
    return ls;

def addData(listData):
    mydb = makeConnectSQL();
    mycursor = mydb.cursor()
    sql = "INSERT INTO MCQData (Problem_Slug, Description, HashMap_Frequency) VALUES (%s, %s, %s)"
    val = []
    for i in listData:
        val.append(tuple(i))
    mycursor.executemany(sql, val)
    mydb.commit()
    print(mycursor.rowcount, "was inserted.") 
    
def sortSlugByPercentage(data):
    sorted_list = sorted(data, key=lambda x:x[1], reverse = True)
    return sorted_list

def getHashMapFrequency(description):
    #Lower case
    description = description.lower()
    
    #Remove special keywords
    description = re.sub('[^a-zA-Z0-9 \n]', '', description)
    
    #Removing stop words
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(description)
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    filtered_sentence = []
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)
            
    #Hashmap of frequency
    hashmap = defaultdict(lambda: 0)
    for i in filtered_sentence:
        hashmap[i] = hashmap[i]+1;
    
    return dict(hashmap)
    
def plagiarismWithFrequency(data, newData):
    ans = [];
    newDataHash = getHashMapFrequency(newData)
    for i in range(len(data)):
        dataHash = eval(data["HashMap_Frequency"][i])
        total = 0;
        common = 0;
#         total2 = sum(list(dataHash.values()))
        
        #get count of common word and total words
        for j in newDataHash.keys():
            total+=newDataHash[j];
            try:
                common+=min(newDataHash[j], dataHash[j]);
            except:
                pass;
            
        #Add slug with percentage of common words
#         total = max(total1, total2)
        if round((common/total)*100, 2)>=20:
            ans.append([data["Problem_Slug"][i], round((common/total)*100, 2)]);
    return sortSlugByPercentage(ans)

def plagiarismWithPysimilar(data, slugList, newData):
    ans = [];
    for i in slugList:
        percentage = compare(data[data["Problem_Slug"]==i[0]]["Description"].to_string(index = False), newData)
        if percentage*100>=65:
            ans.append([i[0], round(percentage*100, 2)]);
        
    return sortSlugByPercentage(ans);
    
```

```python
if __name__ == "__main__":
    excelData = pd.read_excel("AngularFile.xlsx")
    excelData = excelData.astype({'Correct_Answer':'string'})
    excelData["newDescription"] = excelData["Description"]+excelData["Correct_Answer"]
    sqlData = pd.DataFrame(getdata(), columns = ['Problem_Slug', 'Description', 'HashMap_Frequency'])
    newData = [];
    slugs = set(sqlData["Problem_Slug"])
    repeat = [];
    internalPlag = []
    for i in range(len(excelData)):
        ansByHash = (plagiarismWithFrequency(sqlData, excelData["newDescription"][i]))[:30]
        ansByPySimilar = plagiarismWithPysimilar(sqlData, ansByHash, excelData["newDescription"][i])
        internalPlag.append([excelData["Problem_Slug"][i], excelData["newDescription"][i], str(ansByPySimilar)])
        if excelData["Problem_Slug"][i] not in slugs:
            slugs.add(excelData["Problem_Slug"][i])
            newData.append([excelData["Problem_Slug"][i],excelData["newDescription"][i], str(getHashMapFrequency(excelData["newDescription"][i]))])
        else:
            repeat.append([i, excelData["Problem_Slug"][i]])
    
    addData(newData);
```

```python
dataPlag = pd.DataFrame(internalPlag, columns = ["Problem_Slug", "Description", "Plag_Slug_Percentage"])
dataPlag.tail(30)
```
